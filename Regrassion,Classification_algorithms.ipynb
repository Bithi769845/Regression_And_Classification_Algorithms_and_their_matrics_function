```python
df=pd.read_csv('/content/gdrive/MyDrive/Dataset/ddos_dataset_sdn.csv')
df
```

```python
#Handle null values
Mean = df.rx_kbps.mean()
Mean
```

```python
df.rx_kbps = df.rx_kbps.fillna(Mean)
```

```python
df.isnull().sum()
```

```python
Mean = df.tot_kbps.mean()
Mean
df.tot_kbps = df.tot_kbps.fillna(Mean)
```

```python
df.isnull().sum()
```

```python
#One Hot Encoding
dummy_variables = pd.get_dummies(df['Protocol'],drop_first=True)
dummy_variables.head()
new_df=df.drop('Protocol',axis=1)
new_df.head()
df=pd.concat([new_df,dummy_variables],axis=1)
df.head()
```

```python
dummy_variables = pd.get_dummies(df['src'],drop_first=True)
dummy_variables.head()
new_df=df.drop('src',axis=1)
new_df.head()
df=pd.concat([new_df,dummy_variables],axis=1)
df.head()
```

```python
dummy_variables = pd.get_dummies(df['dst'],drop_first=True)
dummy_variables.head()
new_df=df.drop('dst',axis=1)
new_df.head()
df=pd.concat([new_df,dummy_variables],axis=1)
df.head()
```

```python
df=df.drop(['packetins'],axis=1)
```

```python
x=df.drop(['label'],axis=1)
x
```

```python
y=df['label']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=.3,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
Rclf = RandomForestClassifier()
Rclf.fit(xtrain, ytrain)
scr=Rclf.score(xtest,ytest)
print("Scr:",scr)
pre=Rclf.predict(xtest)
print("pre:",pre)
```

# **Matrics**

```python
#Confusion_matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,pre)
print("confusion matrix",cm)
TN = cm[0, 0]
FP = cm[0, 1]
FN = cm[1, 0]
TP = cm[1, 1]
print("True Positives (TP):", TP)
print("False Positives (FP):", FP)
print("False Negatives (FN):", FN)
print("True Negatives (TN):", TN)
#accuracy_score
from sklearn.metrics import accuracy_score
ACCURACY_SCORE = accuracy_score(ytest, pre)
print("Accuracy Score:",ACCURACY_SCORE)
#precision_score
#precision_score
from sklearn.metrics import precision_score
Precision=precision_score(ytest,pre)
print("Precision:",Precision)
#recall_score
from sklearn.metrics import recall_score
Recall=recall_score(ytest,pre)
print("Recall:",Recall)
#Sensitivity
Sensitivity=recall_score(ytest, pre)
print("Sensitivity:",Sensitivity)
#Specificity
Specificity = recall_score(ytest, pre, pos_label=0)
print("Specificity:",Specificity)
#f1_score
from sklearn.metrics import f1_score
F1_SCORE= f1_score(ytest,pre)
print("F1_SCORE:",F1_SCORE)
#Training Time
import time
start_time = time.time()
Rclf.fit(xtrain, ytrain)
end_time = time.time()
training_time = end_time - start_time
print("Training Time: {:.2f} seconds".format(training_time))
#Test Time
start_time = time.time()
Rclf.predict(xtest)
end_time = time.time()
testing_time = end_time - start_time
print("Testing Time: {:.2f} seconds".format(testing_time))
#True Positive Rate (TPR)
TPR = TP / (TP + FN)
print("TPR:",TPR)
#False Positive Rate (FPR)
FPR=FP / (FP + TN)
print("FPR:",FPR)
#Balanced Accuracy (BACC)
from sklearn.metrics import balanced_accuracy_score
BACC= balanced_accuracy_score(ytest,pre)
print("BACC:",BACC)
#ERROR RATE
error_rate = 1 - accuracy_score(ytest,pre)
print("Error_Rate:",error_rate)
#training_accuracy
pre=Rclf.predict(xtrain)
training_accuracy = accuracy_score(ytrain,pre)
print("Training_Accuracy:",training_accuracy)
#test_accuracy
pre=Rclf.predict(xtest)
test_accuracy = accuracy_score(ytest, pre)
print("Test_Accuracy:",test_accuracy)
#cohen_kappa_score
from sklearn.metrics import cohen_kappa_score
kappa = cohen_kappa_score(ytest,pre)
print(f"Cohen's Kappa: {kappa:.5f}")
# Calculate observed accuracy (Po)
Po = np.sum(np.diag(cm)) / np.sum(cm)
#cm is the confusion matrix
print(f"Observed Accuracy (Po): {Po:.5f}")
# Calculate expected accuracy (Pe)
total_samples = np.sum(cm)
total_agreements = np.sum(np.sum(cm, axis=0) * np.sum(cm, axis=1))
Pe = total_agreements / (total_samples * total_samples)
print(f"Expected Accuracy (Pe): {Pe:.5f}")

```
